```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
#faz a leitura dos dados
dados = pd.read_csv('dados.csv', sep = ';')
dados.shape
```

```python
#exibe o cabecalho da planilha
dados.head()
```

```python
#me mostra se existe algum dado nulo na tabela
dados.isnull().sum()
```

```python
# me da uma breve descricao a respeito da coluna MUNICIPIO
dados['MUNICIPIO'].describe()
```

```python
# após analize vejo que alguns municipios estao duplicados, faço o tratamento para tirar as duplicatas
dados = dados.drop_duplicates(subset='MUNICIPIO', keep='first')
agrupado = dados.groupby(['MUNICIPIO']).size()
agrupado
```

```python

p = srn.histplot(dados.PIB, bins = 10).set(title = "PIB")
```

```python
p = srn.histplot(dados.VALOREMPENHO,  bins = 10).set(title = "EMPENHO")
```

```python
#exibe os dados a respeito do pib e seus municipios
agrupado = dados.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color = 'blue')
```

```python
##exibe os dados a respeito do ValorEmpenho e seus municipios
agrupado = dados.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'blue')
```

```python
#exibe uma comparação entre o pib e o ValorEmpenho
plt.scatter(dados.VALOREMPENHO, dados.PIB)
```

```python
# nos mostra onde existe uma concentração maior do pib, e alguns outliers
plt.boxplot(dados.PIB, vert = False)
plt.title('PIB')
plt.xlabel('Dados')
```

```python
# nos mostra onde existe uma concentração maior do Valor Empenho, e alguns outliers
plt.boxplot(dados.VALOREMPENHO, vert = False)
plt.title('VALOR EMPENHO')
plt.xlabel('Dados')
```
